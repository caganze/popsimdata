```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import ascii, fits
from astroquery.vizier import Vizier
from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.table import Table
from astropy.coordinates import Angle
import glob
import splat
%matplotlib inline
```

```python
FOLDER= '/volumes/LaCie/LodieuTemplates/'
```

```python

Vizier.ROW_LIMIT = -1 
catalogs = Vizier.get_catalogs('J/A+A/628/A61')
```

```python
catalogs[0].write('/users/caganze/lodieu2019_sources.csv')
```

```python
cat=catalogs[0].to_pandas()
```

```python
cat
```

```python
#simbad crossmn
simbad_crossmatch=ascii.read(FOLDER+'/lodieu2019_sources_simbad.csv')
gaia_crossmatch= ascii.read(FOLDER+'/lodieu2019_sources_gaia.csv')
sdss_crossmatch= ascii.read(FOLDER+'/lodieu2019_sources_sdss.csv')
```

```python
gaia_crossmatch.to_pandas()
```

```python
spectra=[]
for f in glob.glob(FOLDER+'/*.dat.fits'):
        name=f.split('/')[-1].split('.dat.fits')[0]
        name=name.split('J_A+A_628_A61_sp_')[-1]
        df=Table(fits.open(f)[1].data).to_pandas()
        s=splat.Spectrum(wave=df['lambda'].values/1000,\
                         flux= df.Flux.values, noise=np.zeros(len(df.Flux)),
                        name=name)
        spectra.append(s)
```

```python
data={'spectra': spectra,
     'gaia': gaia_crossmatch,
     'simbad': simbad_crossmatch,
      'sdss': sdss_crossmatch,
     'original':cat}
```

```python
np.save('/Volumes/LaCie/popsimsdata/lodieu_templates.npy', [data], allow_pickle=True)
```
