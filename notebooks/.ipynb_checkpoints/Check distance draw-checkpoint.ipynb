```python
#test the model class on JSWT simulations 
import numpy as np
import pandas as pd
#import splat
#import popsims
from astropy.coordinates import SkyCoord, Galactic
#from popsims.galaxy import Pointing, volume_calc, create_pop
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
#sns.set_style("darkgrid", {"axes.facecolor": ".95"})

#plt.style.use('fivethirtyeight')
#plt.style.use('dark_background')

from  matplotlib.colors import Normalize
import astropy.units as u
import popsims
import matplotlib
from tqdm import tqdm
from tqdm import tqdm
import matplotlib as mpl 
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['xtick.labelsize'] =20
mpl.rcParams['ytick.labelsize'] = 20
mpl.rcParams['legend.fontsize'] = 20
mpl.rcParams['figure.titlesize'] = 20

from popsims.galaxy import Disk, Halo, transform_tocylindrical

%matplotlib inline
```

```python
ls= np.random.uniform(-180, 180, 1000)
bs= np.random.uniform(-90, 90, 1000)

s= SkyCoord(l=ls*u.degree, b=bs*u.degree, frame='galactic')
```

```python
disk= Disk()
```

```python
#it should return the same thing over and over
samples0=disk.sample_distances(ls[0], bs[0], 0.1, 50000, 1e6)
#increase distance steps for more accruacy
samples1=disk.sample_distances(ls[0], bs[0], 0.1, 10000, 1e6, dsteps=5000)
```

```python
fig, ax=plt.subplots()
_= plt.hist(samples0, bins=100, histtype='step')
_= plt.hist(samples1, bins=100,  histtype='step')
```

```python
halo= Halo()
samples2=halo.sample_distances(ls[0], bs[0], 0.1, 50000, 1e6)
#increase distance steps for more accruacy
samples3=halo.sample_distances(ls[0], bs[0], 0.1, 10000, 1e6, dsteps=5000)
fig, ax=plt.subplots()
_= plt.hist(samples2, bins=100, histtype='step')
_= plt.hist(samples3, bins=100,  histtype='step')
```

```python
disk1= Disk(H=300, L=2600)
disk2= Disk(H=900, L=3600)
```

```python
tot=disk1 +0.12*disk2+(1/400)*halo
```

```python
samples3=tot.sample_distances(ls[0], bs[0], 0.1, 10000, 1e6)
samples2=disk1.sample_distances(ls[0], bs[0], 0.1, 10000, 1e6)
fig, ax=plt.subplots()
_= plt.hist(samples2, bins=100, histtype='step')
_= plt.hist(samples3, bins=100,  histtype='step')
```

```python
#draw from random pointings, color them by stellar density
fig, ax=plt.subplots(figsize=(8, 6))
for sx in tqdm(s):
    l= sx.l.radian
    b= sx.b.radian
    ds= tot.sample_distances(l, b, 0.1, 100000, 1e5, dsteps=1000)
    r, z= transform_tocylindrical (l, b, ds)
    ax.scatter(r, z, s=.1, c=np.log10(tot.stellar_density(r, z)), cmap='cubehelix', alpha=0.05)
    
ax.set(xlabel='r (pc)', ylabel='z (pc)')
```

```python
tot2=disk1 +0.05*disk2+(1/400)*halo
```

```python
#draw from random pointings, color them by stellar density
fig, ax=plt.subplots(figsize=(8, 6))
for sx in tqdm(s):
    l= sx.l.radian
    b= sx.b.radian
    ds= tot2.sample_distances(l, b, 0.1, 100000, 1e5, dsteps=1000)
    r, z= transform_tocylindrical (l, b, ds)
    ax.scatter(r, z, s=.1, c=np.log10(tot2.stellar_density(r, z)), cmap='cubehelix', alpha=0.1)
    
ax.set(xlabel='r (pc)', ylabel='z (pc)')
```
