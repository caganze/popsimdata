```python
#imports 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import popsims
from popsims.simulator import Population, pop_mags
from popsims.galaxy import Disk, Halo, Uniform
from popsims.plot_style import  plot_style
from popsims import random_angles

plot_style()

import warnings
warnings.filterwarnings("ignore")
```

```python
p=Population(evolmodel= 'burrows1997',
              imf_power=-0.6,
              binary_fraction=0.2,
              age_range=[8, 14],
              mass_range=[0.01, .1],
             nsample=1e5)
```

```python
p.simulate()
```

```python
g=Uniform()
g1= Disk()
```

```python
l, b=random_angles(10)
```

```python
p.add_distances(g, l, b, 0.1,  1000, dsteps=5000)
```

```python
p.__dict__.keys()
```

```python
fig, ax=plt.subplots()
_= plt.hist(p.distance, bins=32)
```

```python
p.add_distances(g1, l, b, 0.1,  1000, dsteps=5000)
```

```python
fig, ax=plt.subplots()
_= plt.hist(p.distance, bins=32)
```
