```python
def read_model_file(f):
    from astropy.io import ascii
    df=ascii.read(f, fast_reader=False,  names=['wavelength', 'surface_flux'] ).to_pandas()
    
    return df
def get_model_parameters(f):
    teff=float((f.split('/')[-1]).split('_')[1].split('t')[-1])
    logg= float((f.split('/')[-1]).split('_')[2].split('g')[-1])
    z= float((f.split('/')[-1]).split('_')[3].split('z')[-1])
    return {'filename': f, 'teff': teff, 'logg': logg, 'z':z}
```

```python
#unmute for metal-poor models
FOLDER= '/users/caganze/research/splat/resources/SpectralModels/drift/SED/*g5.00*'
fls_all=glob.glob(FOLDER)
params=pd.DataFrame.from_records([get_model_parameters(f) for f in fls_all])
fls_drift = np.array(fls_all)[params.z < -0.5]


#for dwarfs
FOLDER= '/users/caganze/research/splat/resources/SpectralModels/btcond/SED/*g5.00*'
fls_cond=glob.glob(FOLDER)
```

```python
def interpolate_radius(teff, logg, model='burrows1997'):
    model_filename=popsims.DATA_FOLDER+'/models/evolutionary/'+model.lower()+'.csv'
    evolutiomodel=pd.read_csv( model_filename)

    #use the full cloud treatment for saumon models
    if model=='saumon2008':
         evolutiomodel=evolutiomodel[evolutiomodel.cloud=='hybrid']
 
    #make age, teff, mass logarithm scale
    valuest=evolutiomodel.temperature.values
    valueslogg=evolutiomodel.gravity.values
    valuesr= evolutiomodel.radius.values
    #valueslumn=evolutiomodel.luminosity.values

    #valuesm=np.log10(evolutiomodel.mass.values)
    #valuesag=np.log10(evolutiomodel.age.values)
    evolpoints=np.array([valuest, valueslogg ]).T

    return griddata(evolpoints, valuesr , (teff, logg), method='linear')

def compute_mag_from_teff(teff):
    return teff_to_mag_kirkpatrick(teff, nsample=1000)

def compute_mag_from_type(spt):
    return spe.typeToMag(spt, '2MASS J')[0]
```

```python
def make_spectrum_from_model(f,  teff, logg):
    FUNIT=u.erg/(u.cm**2*u.micron*u.second)
    df=read_model_file(f)
    #print (df)
    sp= splat.Spectrum(wave=df.wavelength.values*u.micron,\
                       flux=df.surface_flux.values*FUNIT, \
                       noise=1e-10*(df.surface_flux.values)*FUNIT,\
                       name=f.split('/')[-1],
                        flux_unit=FUNIT)
    #scale spectrum by radius
    sp.trim([0.01, 3.0])
    radius=  interpolate_radius(teff, logg, model='baraffe2003')
    scalef= (radius*u.Rsun.to(u.pc)/10)**2
    sp.scale(scalef)
    return sp

def make_line_spectrum_from_model(f,  teff, logg):
    #FUNIT=u.Watt/(u.m**2*u.m)
    FUNIT=u.erg/(u.cm**2*u.micron*u.second)
    df=read_model_file(f)
    #print (df)
    sp= splat.Spectrum(wave=df.wavelength.values*u.micron,\
                       flux=df.surface_flux.values*FUNIT, \
                       noise=1e-10*(df.surface_flux.values)*FUNIT,\
                       name=f.split('/')[-1],
                       flux_unit=FUNIT)
    #scale spectrum by radius
    sp.trim([0.01, 3.0])
    radius=  interpolate_radius(teff, logg, model='baraffe2003')
    scalef= (radius*u.Rsun.to(u.pc)/10)**2
    sp.scale(scalef)
    return sp
```

```python
drift_df= pd.DataFrame.from_records([get_model_parameters(f) for f in fls_drift])
btcond_df=pd.DataFrame.from_records([get_model_parameters(f) for f in fls_cond])
```

```python
spectra_drift= [make_spectrum_from_model(f, teff, logg) for f, teff, logg in zip(drift_df.filename.values,\
                                                                              drift_df.teff.values,\
                                                                           drift_df.logg.values) ]
btcond_spectra=[make_spectrum_from_model(f, teff, logg) for f, teff, logg in zip(btcond_df.filename.values,\
                                                                              btcond_df.teff.values, \
                                                                                 btcond_df.logg.values) ]
```
