```python
def get_abs_mag(mag, d, mag_err, d_err ):
    ds= np.random.normal(d, d_err, 1000)
    mags= np.random.normal(mag, mag_err, 1000)
    vals= mags-5*np.log10(ds/10.0)
    return np.nanmedian(vals), np.nanstd(vals)
```

```python
c='PS1gmag'
t['f_{}'.format(c)].apply(lambda x: x.strip()).values
```

```python
#remove binaries
bools=[]
bools.append(np.logical_and(t.Young.apply(lambda x: x.strip()) !='Y', t.Binary.apply(lambda x: x.strip()) !='Y').values)

#remove based on photometry flags
#rejected after visual inspection
#rejection flags
columns=['PS1gmag', 'PS1rmag', 'PS1imag', 'PS1zmag', 'PS1ymag', 'Jmag', 'Hmag', 'Ksmag',  'W1mag','W2mag','W3mag']
for c in columns:
    try:
        bools.append((t['f_{}'.format(c)].apply(lambda x: x.strip()) =='').values)
    except:
        pass
```

```python
cut= (t.e_Dist.values/t.Dist.values) < .1
#add 2 mass flag
bools[-1]= np.array(cut)
bools1= np.logical_and.reduce(bools)
```

```python
cut
```

```python
df=t[bools1]
```

```python
ntypes= df.SpT.apply(lambda x: splat.typeToNum(x))
```

```python
abs_mags={}
masks={}
for c in columns:
    #it looks like abs mags should 
    fig, ax=plt.subplots(figsize=(8, 6))
    m= df[[c, 'Dist', 'e_{}'.format(c), 'e_Dist']].apply(lambda x: get_abs_mag(*x), axis=1)
    ax.errorbar(ntypes, np.vstack(m)[:,0], fmt='.', yerr=np.vstack(m)[:,1])
    abs_mags[c]=np.vstack(m)[:,0]
    abs_mags[c+'_er']= np.vstack(m)[:,1]
    mask= np.vstack(m)[:,0]<3
    masks[c]=mask
    ax.axhline(1, color='k')
    ax.set(xlabel='Spectral Type', ylabel=c)
    ax.set(xlabel='Spectral Type', ylabel=c)
```

```python
len()
```

```python
for c in columns:
    fig, ax=plt.subplots(figsize=(8, 6))
    mask= masks[c]
    df['absolute_{}'.format(c)]= abs_mags[c]
    df['absolute_{}_er'.format(c)]=  abs_mags[c+'_er']
    ax.errorbar(ntypes[~mask], abs_mags[c][~mask], yerr=np.vstack(m)[:,1][~mask], fmt='o')
    ax.set(xlabel='Spectral Type', ylabel=c)
```

```python
for c in columns: 
    df['use_mag_{}'.format(c)] = (~masks[c]).astype(int)
```

```python
df['NumType']= ntypes
```

```python
df.reset_index(drop=True).to_pickle('/Volumes/LaCie/Best2018_masked.csv.gz')
```
