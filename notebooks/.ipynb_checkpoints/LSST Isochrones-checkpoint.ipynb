```python
fig, ax=plt.subplots()
ax.scatter(dp0_data.mag_r-dp0_data.mag_z, dp0_data.mag_g-dp0_data.mag_r, s=1, color='k')
c=ax.scatter(bckg.abs_LSST_R-bckg.abs_LSST_Z,
             bckg.abs_LSST_G-bckg.abs_LSST_R, 
             c=np.log10(bckg.age), s=1., marker='+', cmap='coolwarm')
plt.colorbar(c, label='Age')
ax.set(xlim=[-1, 4], ylim=[-1, 5], xlabel='r-z', ylabel='g-r')
```
