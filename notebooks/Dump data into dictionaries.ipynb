```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
#binaries
```

```python
df=pd.read_csv('/volumes/lacie/popsimsdata/popsims/binary_lookup_table.csv.gz')

xgrid= np.arange(15, 42)
systems=[]
for sp in xgrid:
    for ssec in xgrid:
        query=df.query('prim == {} & sec =={}'.format(sp, ssec))
        systems.append({'prim': sp,
                      'sec': ssec,
                      'sys': np.nanmedian(query.sys.values),
                      'sys_unc': (np.nanmedian(query.sys.values)**2+np.nanmedian(query.sys_unc.values)**2)**0.5})
    
```

```python

dfm=pd.DataFrame.from_records(systems).dropna()

import json

with open('/users/caganze/research/popsimsdata/binaries.txt', 'w') as file:
     file.write(json.dumps(dfm.to_dict(orient='list'))) # use `json.loads` to do the reverse
```

```python
plt.scatter(dfm.prim, dfm.sys)
```
