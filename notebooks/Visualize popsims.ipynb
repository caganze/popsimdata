```python
import pandas as pd
import numpy as np
import popsims

import matplotlib.pyplot as plt
import glob
from popsims.plot_style import  plot_style
from popsims.galaxy import Disk, Halo
plot_style()
%matplotlib inline
```

```python
#define a galaxy density model and draw distances in a given direction
disk= Disk(H=300, L=2600)
tdisk=Disk(H=1000, L=3600)
halo= Halo()
model= disk+0.12*tdisk+0.0025*halo
```

```python
ax=model.plot_countours(log=True, rmax=50_000, zmax=10000, zmin=-10000)
#ax.set()
```
